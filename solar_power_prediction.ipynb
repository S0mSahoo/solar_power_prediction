```python
#Plot distribution of power
plt.figure(figsize=(10, 6))
sns.histplot(df['generated_power_kw'], bins=30, kde=True)
plt.title('DIstribution of Generated Power in kW')
plt.xlabel('Generated Power (kW)')
plt.ylabel('Frequency')
plt.show()
```

### Exploratory Data Analysis(Week 2)

```python
#Histograms for remaining columns

df[df.columns[:9]].hist(bins=30, figsize=(15, 10))
# df.hist(bins=30, figsize=(15, 10))
plt.show()
```

```python
df[df.columns[10:20]].hist(bins=30, figsize=(15,15))
plt.show()
```

```python
#Bivariate Analysis using ScatterPlot
#ScatterPlot with Target Feature
plt.figure(figsize=(15, 30))
for i, column in enumerate(df.columns):
    plt.subplot(7, 3, i+1)
    plt.scatter(df[column], df['generated_power_kw'])
    plt.title(f'{column} vs Generated power in kW')
    plt.xlabel(column)
    plt.ylabel('Generated power in kW')
plt.tight_layout()
plt.show()
```
